```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
```

```python
#Funcin 1
def graficar_recta(m, b):
    x = np.linspace(-10, 10, 400)
    y = m * x + b

    plt.figure(figsize=(6, 6))
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.plot(x, y, label=f'y = {m}x + {b}', color='purple')
    plt.legend()
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.show()

interact(graficar_recta, m=widgets.FloatSlider(min=-10, max=10, step=0.5, value=1),
                         b=widgets.FloatSlider(min=-10, max=10, step=0.5, value=0));
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

def graficar_reflejos(m, b):
    x = np.linspace(-10, 10, 500)
    y = m * x + b

    # Reflejos
    y_reflejo_x = -y          # Reflejo en el eje X
    y_reflejo_y = m * (-x) + b  # Reflejo en el eje Y
    y_reflejo_ambos = -(m * (-x) + b)

    plt.figure(figsize=(6, 6))
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.grid(True, linestyle='--', alpha=0.7)

    # Graficar solo los reflejos
    plt.plot(x, y_reflejo_x, label=f'Reflejo en eje X', color='red')
    plt.plot(x, y_reflejo_y, label=f'Reflejo en eje Y', color='blue')
    plt.plot(x, y_reflejo_ambos, label=f'Reflejo en ambos ejes', color='green', linestyle='dashed')

    plt.legend(loc='upper left')
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.show()

# Deslizadores para m y b
interact(graficar_reflejos, m=widgets.FloatSlider(min=-10, max=10, step=0.5, value=1),
                            b=widgets.FloatSlider(min=-10, max=10, step=0.5, value=0));
```

```python
#Funcin 2
def graficar_parabola(a, b, c):
    x = np.linspace(-10, 10, 500)
    y = a*((-x)**(2))+b*(-x)+c

    # Reflejos
    y_reflejo_x = -y          # Reflejo en el eje X
    y_reflejo_y = a*((-x)**(2))+b*(-x)+c  # Reflejo en el eje Y
    y_reflejo_ambos = -(a*((-x)**(2))+b*(-x)+c)

    plt.figure(figsize=(6, 4))
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.grid(True, linestyle="--", linewidth=0.5)

   # plt.plot(x, y_reflejo_x, label=f'Reflejo en eje X', color='red')
    plt.plot(x, y_reflejo_y, label=f'Reflejo en eje Y', color='blue')
    #plt.plot(x, y, label=f'$y = {a}x^2 + {b}x + {c}$')

    plt.xlim(-10, 10)
    plt.ylim(-20, 20)
    plt.legend()
    plt.show()

interact(graficar_parabola,
         a=widgets.IntSlider(min=-10, max=10, step=1, value=1),
         b=widgets.IntSlider(min=-10, max=10, step=1, value=0),
         c=widgets.IntSlider(min=-10, max=10, step=1, value=0))
```

```python
#Funcin 3
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

def graficar_raiz(a, b, c, e):
    x = np.linspace(-10, 10, 400)
    x = x[x >= c/b] if b != 0 else x  # Restringir dominio para evitar races negativas
    y = a * np.sqrt(b * x - c) + e

    # Reflejos
    y_reflejo_x = -y          # Reflejo en el eje X
    y_reflejo_y = a * np.sqrt(b * (-x) - c) + e  # Reflejo en el eje Y
    y_reflejo_ambos = -(a * np.sqrt(b * (-x) - c) + e)

    plt.figure(figsize=(6, 6))
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.grid(True, linestyle='--', alpha=0.6)
    #plt.plot(x, y, label=f'$y = {a}\sqrt{{{b}x - {c}}} + {e}$', color='purple')
    #plt.plot(x, y_reflejo_x, label=f'Reflejo en eje X', color='red')
    plt.plot(x, y_reflejo_y, label=f'Reflejo en eje Y', color='blue')
    #plt.plot(x,y_reflejo_ambos, label=f'Reflejo de ambos', color='green', linestyle='dashed')


    plt.legend()
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.show()

interact(graficar_raiz,
         a=widgets.FloatSlider(min=-10, max=10, step=0.5, value=1),
         b=widgets.FloatSlider(min=-10, max=10, step=0.5, value=1),
         c=widgets.FloatSlider(min=-10, max=10, step=0.5, value=0),
         e=widgets.FloatSlider(min=-10, max=10, step=0.5, value=0));
```

```python
#Funcin 4
def graficar_seno(a, b):
    x = np.linspace(-10, 10, 400)
    y = a * np.sin(b * x)

    # Reflejos
    y_reflejo_x = -y          # Reflejo en el eje X
    y_reflejo_y = a * np.sin(b * (-x))  # Reflejo en el eje Y
    y_reflejo_ambos = -(a * np.sin(b * (-x)))

    plt.figure(figsize=(6, 6))
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.plot(x, y, label=f'$y = {a} \sin({b}x)$', color='purple')
    plt.plot(x, y_reflejo_x, label=f'Reflejo en eje X', color='red')
    plt.plot(x, y_reflejo_y, label=f'Reflejo en eje Y', color='blue')
    plt.plot(x,y_reflejo_ambos, label=f'Reflejo de ambos', color='green', linestyle='dashed')



    plt.legend()
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.show()

interact(graficar_seno,
         a=widgets.FloatSlider(min=-10, max=10, step=0.5, value=1),
         b=widgets.FloatSlider(min=-10, max=10, step=0.5, value=1));
```

```python
#Funcin 5
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

def graficar_coseno(b):
    x = np.linspace(-10, 10, 400)
    y = np.cos(b * x)

    # Reflejos
    y_reflejo_x = -y          # Reflejo en el eje X
    y_reflejo_y = np.cos(b * (-x))  # Reflejo en el eje Y
    y_reflejo_ambos = -(np.cos(b * (-x)))

    plt.figure(figsize=(6, 6))
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.plot(x, y, label=f'$y = \cos({b}x)$', color='purple')
    plt.plot(x, y_reflejo_x, label=f'Reflejo en eje X', color='red')
    plt.plot(x, y_reflejo_y, label=f'Reflejo en eje Y', color='blue', linestyle='dashed')
    plt.plot(x,y_reflejo_ambos, label=f'Reflejo de ambos', color='pink', linestyle='dashed')

    plt.legend()
    plt.xlim(-10, 10)
    plt.ylim(-1.5, 1.5)
    plt.show()

interact(graficar_coseno,
         b=widgets.FloatSlider(min=-10, max=10, step=0.5, value=1));
```

```python
#Funcin 6
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

def graficar_exponencial(a, b, c):
    x = np.linspace(-10, 10, 400)
    y = a * np.exp(b * x) + c

    # Reflejos
    y_reflejo_x = -y          # Reflejo en el eje X
    y_reflejo_y = a * np.exp(b * (-x)) + c  # Reflejo en el eje Y
    y_reflejo_ambos = -(a * np.exp(b * (-x)) + c)

    plt.figure(figsize=(6, 6))
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.plot(x, y, label=f'$y = {a}e^{{{b}x}} + {c}$', color='purple')
    plt.plot(x, y_reflejo_x, label=f'Reflejo en eje X', color='red')
    plt.plot(x, y_reflejo_y, label=f'Reflejo en eje Y', color='blue', linestyle='dashed')
    plt.plot(x,y_reflejo_ambos, label=f'Reflejo de ambos', color='pink', linestyle='dashed')

    plt.legend()
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.show()

interact(graficar_exponencial,
         a=widgets.FloatSlider(min=-10, max=10, step=0.5, value=1),
         b=widgets.FloatSlider(min=-10, max=10, step=0.5, value=1),
         c=widgets.FloatSlider(min=-10, max=10, step=0.5, value=0));
```

```python
#Funcin 7
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

def graficar_logaritmica(a, b, c):
    x = np.linspace(-10, 10, 400)
    x = x[x * b + c > 0]
    y = a * np.log(b * x + c)

    # Reflejos
    y_reflejo_x = -y          # Reflejo en el eje X
    y_reflejo_y = a * np.log(b * (-x) + c)  # Reflejo en el eje Y
    y_reflejo_ambos = -(a * np.log(b * (-x) + c))

    plt.figure(figsize=(6, 6))
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.plot(x, y, label=f'$y = {a} \\ln({b}x + {c})$', color='purple')
    plt.plot(x, y_reflejo_x, label=f'Reflejo en eje X', color='red')
    plt.plot(x, y_reflejo_y, label=f'Reflejo en eje Y', color='blue',linestyle='dashed')
    plt.plot(x,y_reflejo_ambos, label=f'Reflejo de ambos', color='green', linestyle='dashed')

    plt.legend()
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.show()

interact(graficar_logaritmica,
         a=widgets.FloatSlider(min=-10, max=10, step=0.5, value=1),
         b=widgets.FloatSlider(min=-10, max=10, step=0.5, value=1),
         c=widgets.FloatSlider(min=-10, max=10, step=0.5, value=0));
```
